```python
%%time

# pre-trained models on general text.

model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
# This is a sentence-transformers model: 
# It maps sentences & paragraphs to a 768 dimensional dense vector space
# Our model is intented to be used as a sentence and short paragraph encoder. 
# Given an input text, it ouptuts a vector which captures the semantic information. 
# The sentence vector may be used for information retrieval, clustering or sentence similarity tasks.
# By default, input text longer than 384 word pieces is truncated.
```

```python
%%time 

# Tune the model

embeddings = model.encode(sentences, batch_size = 128, convert_to_numpy = True, show_progress_bar=True)
```

```python
print('rows: ', len(embeddings))
print('columns: ', len(embeddings[1]))
print('sentences: ', len(sentences))
```

```python
%%time

# umap_neighbors: controls how UMAP balances local versus global structure in the data. 
# It does this by constraining the size of the local neighborhood UMAP will look at when attempting 
# to learn the manifold structure of the data. This means that low values of n_neighbors will force UMAP to concentrate 
# on very local structure (potentially to the detriment of the big picture)

umap_neighbors = 20   # 12
# The size of local neighborhood (in terms of number of neighboring sample points) used for manifold approximation


umap_n_components = 12  # 12 dimensions 
# you will get significantly diminishing returns for an embedding dimension larger than the n_neighbors value. 


umap_embeddings_12d = umap.UMAP(n_neighbors=umap_neighbors,
                            n_components=umap_n_components,
                            n_epochs=14000,
                            min_dist=0.1, # The effective minimum distance between embedded points. Smaller values will result in a more clustered/clumped embedding
                            low_memory=False,
                            learning_rate=0.5,
                            verbose=True,
                            metric='cosine',
                            spread=3.0,
                            local_connectivity=2,
                            target_metric='l2').fit_transform(embeddings)   # embeddings mit 1024 dimensions
```

```python
%%time

hdbscan_minimal_cluster_size = 70   # minimal cluster 200 --> EU 118 cluster ,   min cluster 500 --> 63 cluster
hdbscan_min_samples = 4      # larger value of "min_samples" provide more conservative the clustering 
                             # – more points will be declared as noise, and
                             # clusters will be restricted to more dense areas. 





cluster = hdbscan.HDBSCAN(min_cluster_size = hdbscan_minimal_cluster_size,
                          metric='euclidean', # euclidean
                          min_samples = hdbscan_min_samples, # if not specified the default value is equal to min_cluster_size.
                          #core_dist_n_jobs = 10,
                          cluster_selection_epsilon=0.01, # A distance threshold. Clusters below this value will be merged.
                          cluster_selection_method='leaf', # select clusters from the condensed tree: eom, leaf
                          leaf_size=40,
                          algorithm='best').fit(umap_embeddings_12d)  # umap_embeddings_12d     prims_kdtree (10x longer than boruvka_kdtree), boruvka_kdtree, boruvka_balltree (slower), generic (crash)
                          #prediction_data=True)  # umap_embeddings_12d    # embeddings

    
anzahl_clusters = len(pd.Series(cluster.labels_).unique())
print('Number of sentences = ', len(sentences))
print('number of Cluster = ', anzahl_clusters, '\n')

df4 = pd.DataFrame(pd.DataFrame(cluster.labels_).value_counts())   #convert cluster.labels_ to pandas df

# Head of pandas df:

with pd.option_context('display.max_rows', 30, 'display.min_rows', 10, 'display.max_columns', 4, 'display.width', 100):  # Width of the display in characters. default=80
    print (df4.head(20))
```

```python
%%time


umap_data_3d = umap.UMAP(n_neighbors=umap_neighbors,
                         #densmap=True,
                         #dens_lambda=2.0, # default = 2
                         n_components=3,  # Graph in 3D
                         min_dist=0.01,
                         spread=2.0,
                         low_memory=False,
                         n_epochs=2500,
                         learning_rate = 1.0,
                         target_metric='l2',
                         verbose=True,
                         local_connectivity=2,
                         metric='cosine').fit_transform(embeddings)  # cosine  # 1024 dim # umap_embeddings_12d
```

```python
result3d = pd.DataFrame(umap_data_3d, columns=['x', 'y', 'z'])
result3d['sentence'] = sentences
result3d['labels'] = cluster.labels_

with pd.option_context('display.max_rows', 30, 'display.min_rows', 10, 'display.max_columns', 5, 'display.width', 210, 'display.max_colwidth', 150):  # Width of the display in characters. default=80
    print (result3d.head(39))

print('No of Cluster = ', anzahl_clusters)
```
